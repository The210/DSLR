```python
import pandas as pd
import math
import sys
import csv
```

```python
def read_file(csv_file):
    data = []
    with open(csv_file, mode='r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data
```

```python
def check_numeric(column, data):
    had_value = False
    for row in data[1:]:
        if row[column] == "":
            continue
        had_value = True
        try:
            float(row[column])
        except ValueError:
            return False
    return had_value
```

```python
def numeric_dataset(data):
    new_data = []
    for i in range(len(data[0])):
        if (check_numeric(i, data)):
            column_values = [row[i] for row in data]
            new_data.append(column_values)
    return new_data
```

```python
def remove_first_value(data):
    removed_values = []
    for row in data:
        if len(row) > 1:
            removed_values.append(row.pop(0))
    return removed_values
```

```python
def normalize_col(values):
    maxx = float(values[0]) if values[0] != '' else 0.0
    for value in values:
        if value == '':
            value = 0.0
        if abs(float(value)) > maxx:
            maxx = abs(float(value))
    for i, value in enumerate(values):
        if value == '':
            values[i] = 0.0
            continue
        values[i] = float(value) / maxx if maxx != 0 else 0.0
    return values

def normalize(data):
    for i, col in enumerate(data):
        data[i] = normalize_col(col)
    return data
```
